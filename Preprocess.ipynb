```python
df = df.sort_values(by=['director_name'], ascending = True)
```

```python
df.head()
```

```python
network = df[['director_name', 'actor_2_name']]
```

```python
network.head()
```

```python
network.dropna(inplace=True)
```

```python
len(network)
```

```python
ax = network['director_name'].value_counts().plot(kind='bar',
                                    figsize=(14,8),
                                    title="Director")
```

```python
network['director_name'].value_counts()
```

```python
## Creating a list with both directors and actor names
```

```python
director_names = network['director_name'].tolist()
```

```python
actor_names = network['actor_2_name'].tolist()
```

```python
all_names = director_names + actor_names
```

```python
# Creating a dictionary that maps each word to its number of occurances   
mapper = {}
for name in all_names:
    if name not in mapper:
        mapper[name] = 1
    else:
        mapper[name] += 1
```

```python
mapper
```

```python
# Creating a dictionary that maps each name from mapper to a unique integer
# We keep the words that occur at least 20 times
# threshold = 20
name2int = {}
word_number = 0
for name, count in mapper.items():
    name2int[name] = word_number
    word_number += 1
```

```python
name2int
```

```python
'Aaron Hann' in name2int
```

```python
def transform_name(entry):
    if entry in name2int:
        return name2int[entry]
```

```python
network['director_number'] = network['director_name'].apply(transform_name)
```

```python
network.head()
```

```python
network['actor_number'] = network['actor_2_name'].apply(transform_name)
```

```python
network.head()
```

```python
final_network = network[['director_number', 'actor_number']]
```

```python
final_network.head()
```

```python
final_network.reset_index(inplace=True)
```

```python
del final_network['index']
```

```python
final_network.head()
```

```python
final_network.to_csv('Dataset.csv', header=None, index=False)
```

```python
tfile = open('Dataset.txt', 'a')
tfile.write(final_network.to_string(header=False, index=False))
tfile.close()
```
